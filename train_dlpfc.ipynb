```python
import pytorch_lightning as pl
from torch.utils.data import DataLoader

from dataset import ViT_HER2ST, MyDataset
from vis_model import THItoGene

import warnings
warnings.filterwarnings("ignore")

section_id = '151673'
tag = '-dlpfc'
dataset = MyDataset(dataset='dlpfc', train=True, section_id=section_id)
train_loader = DataLoader(dataset, batch_size=1, num_workers=0, shuffle=True)
model = THItoGene(n_genes=1000, learning_rate=1e-5, route_dim=64, caps=20, heads=[16, 8], n_layers=4)
trainer = pl.Trainer(accelerator="gpu", devices=[0], max_epochs=200)
```

```python
trainer.fit(model, train_loader)
trainer.save_checkpoint("model/last_train_" + tag + '_' + str(section_id) + ".ckpt")
```
